```python
1. Сколько различных комитетов из 3 человек можно выбрать из 10 человек?

Решение:

Формула: C(n,k) = n! / (k! * (n-k)!)
Данные:
n = 10 (общее число людей)
k = 3 (число членов комитета)
```

```python
2. Сколько различных 3-значных чисел можно составить из цифр 0, 1, 2, 3, если цифры не могут повторяться?

Решение:

Формула: C(n,k) = n! / (k! * (n-k)!)
Данные:
n = 4 (общее число цифр)
k = 3 (число используемых цифр)
```

```python
3. Сколько различных 3-буквенных слов можно составить из букв слова "дом"?

Решение:

Формула: C(n,k) = n! / (k! * (n-k)!)
Данные:
n = 3 (общее число букв)
k = 3 (число используемых букв)
```

```python
4. Сколько различных 3-значных чисел можно составить из цифр 0, 1, 2, если цифры могут повторяться?

Решение:

Формула: C(n+k-1,k) = (n+k-1)! / (k! * (n-1)!)
Данные:
n = 3 (общее число цифр)
k = 3 (число используемых цифр)
```

```python
6. Сколько различных вариантов выбрать 3 шара из 5, если шары могут повторяться?

Решение:

Формула: C(n+k-1,k) = (n+k-1)! / (k! * (n-1)!)
Данные:
n = 5 (общее число шаров)
k = 3 (число выбранных шаров)
```
